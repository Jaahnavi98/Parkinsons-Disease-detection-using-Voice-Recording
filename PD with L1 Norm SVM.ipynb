```python
# Calculating Performance metrics - With all features

from sklearn.metrics import classification_report,confusion_matrix

CM1 = confusion_matrix(y_test,y_pred)

TN1 = CM1[0][0]
FN1 = CM1[1][0]
TP1 = CM1[1][1]
FP1 = CM1[0][1]

TPR1 = TP1/(TP1+FN1)
TNR1 = TN1/(TN1+FP1)
PPV1 = TP1/(TP1+FP1)
NPV1 = TN1/(TN1+FN1)
FPR1 = FP1/(FP1+TN1)
FNR1 = FN1/(TP1+FN1)
FDR1 = FP1/(TP1+FP1)

ACC1 = (TP1+TN1)/(TP1+FP1+FN1+TN1) 

print("PERFORMANCE METRICS WITH ALL FEATURES")
print(CM1)
print(classification_report(y_test,y_pred))
print("Sensitivity",TPR1)
print("Specificity",TNR1)

#The classification report shown above says that the model has predicted 
#88% for patients who are not having parkinsons disease 
#and a accuuracy of 83% for person's who are having parkinsons disease. 
#Also the confusion matrix shows that fourteen values have been mis-classified.
```
